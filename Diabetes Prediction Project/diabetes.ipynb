# Modeling

## Splitting Data

```python
from sklearn.model_selection import train_test_split

X = df.drop('Outcome', axis=1)
y = df['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
print(X)
print(X_train)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr_y_pred = lr.predict(X_test)

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

lr_acc = accuracy_score(y_test, lr_y_pred)
lr_conf = confusion_matrix(y_test, lr_y_pred)
lr_class = classification_report(y_test, lr_y_pred)

print('Accuracy Score: ', lr_acc)
print('Confusion Matrix: \n', lr_conf)
print('Classification Report: \n', lr_class)
```

## Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
dt_y_pred = dt.predict(X_test)

dt_acc = accuracy_score(y_test, dt_y_pred)
dt_conf = confusion_matrix(y_test, dt_y_pred)
dt_class = classification_report(y_test, dt_y_pred)

print('Accuracy Score: ', dt_acc)
print('Confusion Matrix: \n', dt_conf)
print('Classification Report: \n', dt_class)
```

## Random Forest Regression Model

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
rf_y_pred = rf.predict(X_test)

rf_acc = accuracy_score(y_test, rf_y_pred)
rf_conf = confusion_matrix(y_test, rf_y_pred)
rf_class = classification_report(y_test, rf_y_pred)

print('Accuracy Score: ', rf_acc)
print('Confusion Matrix: \n', rf_conf)
print('Classification Report: \n', rf_class)
```
