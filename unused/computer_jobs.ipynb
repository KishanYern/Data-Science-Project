```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
warnings.filterwarnings("ignore", category=FutureWarning)
from matplotlib import rcParams
rcParams["figure.figsize"] = 12, 6
sns.set_style('darkgrid')
```

## Importing data set

```python
df = pd.read_excel('datasets/uncleaned_computer_jobs_original.xlsx')
df
```

## Cleaning Data

```python
df = df.drop(columns = ['index'], axis = 1)
df
```

```python
df.describe()
```

```python
df.head()
```

```python
df['Company Name'] = df['Company Name'].str.extract(r'([^\n]+)') #extracting the company name from the company name column
df
```

```python
df.drop_duplicates(subset='Company Name', inplace=True)
df.reset_index(drop=True, inplace=True)  # Reset index and drop the old index

df
```

```python
split_columns = df['Salary Estimate'].str.split('-') #splitting the salary estimate column into min and max salary
df['Min Salary'] = split_columns.str[0].str.extract('(\d+)').astype(int)
df['Max Salary'] = split_columns.str[1].str.extract('(\d+)').astype(int)
df = df.drop(columns = ['Salary Estimate'], axis = 1)
df
```

```python
df = df.drop(columns = ['Job Description'], axis = 1) #dropping the job description column
df
```

```python
df['Size'] = df['Size'].str.replace(' employees', '') #removing the employees from the size column
df['Size'] = df['Size'].str.replace(' to ', '-') #replacing the ' to ' with a dash
df
```

```python
df = df[df['Rating'] != -1] #checking for the rows with -1 rating's and removing them
df.reset_index(drop=True, inplace=True) #resetting the index
df
```

```python
split_locations = df['Location'].str.split(', ') #splitting the location column into city and state
df['Location_City'] = split_locations.str[0]
df['Location_State'] = split_locations.str[1]

split_headquarters = df['Headquarters'].str.split(', ') #splitting the headquarters column into city and state
df['Headquarters_City'] = split_headquarters.str[0]
df['Headquarters_State/Country'] = split_headquarters.str[1]

df = df.drop(columns = ['Headquarters'], axis = 1)
df = df.drop(columns = ['Location'], axis = 1)
df
```

```python
df['Revenue'] = df['Revenue'].str.replace('Unknown / Non-Applicable', 'N/A') #replacing the Unknown / Non-Applicable with Unknown/Non-Applicable
df['Revenue'] = df['Revenue'].str.replace(' (USD)', '') #removing the $ from the revenue column
df['Revenue'].unique()
```

```python
df = df.drop(columns = ['Founded', 'Competitors'], axis = 1) #dropping the founded and competitor column
df
```

## Models

# Splitting the data

```python
X = df.drop(['Min Salary', 'Max Salary'], axis=1)
y_min = df['Min Salary']
y_max = df['Max Salary']

X_encoded = pd.get_dummies(X) # One-hot encode the categorical variables

from sklearn.model_selection import train_test_split
X_train, X_test, y_min_train, y_min_test, y_max_train, y_max_test = train_test_split(
    X_encoded, y_min, y_max, test_size=0.2, random_state=42)
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

model_min = LinearRegression() # Linear model for Min Salary
model_min.fit(X_train, y_min_train)

model_max = LinearRegression() # Linear model for Max Salary
model_max.fit(X_train, y_max_train)

# Evaluate the models
y_min_pred = model_min.predict(X_test)
y_max_pred = model_max.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

mae_min = mean_absolute_error(y_min_test, y_min_pred)

print(f'Min Salary MAE: {mae_min}')
```

```python
mae_max = mean_absolute_error(y_max_test, y_max_pred)

print(f'Max Salary MAE: {mae_max}')
```

```python
from sklearn.metrics import r2_score

r2_min = r2_score(y_min_test, y_min_pred)

print(f'Min Salary R2: {r2_min}')
```

```python
r2_max = r2_score(y_max_test, y_max_pred)

print(f'Max Salary R2: {r2_max}')
```

# kNN Model

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor

df.fillna(method='ffill', inplace=True)
df.replace('N/A', 'Unknown', inplace=True)

# Select features and target variables
X = df.drop(['Min Salary', 'Max Salary'], axis=1)
y_min = df['Min Salary']
y_max = df['Max Salary']

# One-hot encode categorical variables
X_encoded = pd.get_dummies(X, drop_first=True)

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

# Split data into training and test sets
X_train, X_test, y_min_train, y_min_test, y_max_train, y_max_test = train_test_split(
    X_scaled, y_min, y_max, test_size=0.2, random_state=42
)

# Initialize and train a KNN regressor for Min Salary
model_min = KNeighborsRegressor(n_neighbors=5)  # Using 5 neighbors as an example
model_min.fit(X_train, y_min_train)

# Initialize and train a KNN regressor for Max Salary
model_max = KNeighborsRegressor(n_neighbors=5)
model_max.fit(X_train, y_max_train)

# Evaluate the models
y_min_pred = model_min.predict(X_test)
y_max_pred = model_max.predict(X_test)

mae_min = mean_absolute_error(y_min_test, y_min_pred)
mae_max = mean_absolute_error(y_max_test, y_max_pred)

r2_min = r2_score(y_min_test, y_min_pred)
r2_max = r2_score(y_max_test, y_max_pred)

print(f'Min Salary MAE: {mae_min}')
print(f'Max Salary MAE: {mae_max}')
print(f'Min Salary R2 Score: {r2_min}')
print(f'Max Salary R2 Score: {r2_max}')
```
